## Part 2: Use K-means clustering to identify the optimal number of clusters. Experiment with different values of K and use metrics such as the elbow method and silhouette score to evaluate the performance of the clustering.

```python
#adding necessary tools for this part
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

#Assuming df2 as data or X
X = df2
```

**Elbow method**

for Elbow method once we use distortion and another time we use inertia.<br>
a quick brief about distortion and inertia :<br>
Distortion :<br>
It is calculated as the average of the squared distances from the cluster centers of the respective clusters to each data point. Typically, the Euclidean distance metric is used.<br>
$$ 
Distortion = \frac{1}{n} \times \sum \lVert point - centroid \rVert ^2
$$
Inertia :<br>
It is the sum of the squared distances of samples to their closest cluster center.<br>
$$ 
Inertia = \sum \lVert point - centroid \rVert ^2
$$

```python
#adding values of distortions and inertias and then calculating mapping for each
distortions = []
inertias = []

K = range(2, 21)
 
for num_clusters in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters = num_clusters).fit(X)
    kmeanModel.fit(X)
 
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / X.shape[0])
    inertias.append(kmeanModel.inertia_)
 
```

```python
#plotting Elbow Method using Distortion

plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
#plotting Elbow Method using Inertia

plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

**silhouette score**

$$
S(i) = \frac{b(i)-a(i)}{max(a(i),b(i))}
$$
<br>
S(i) is the silhouette coefficient of the data point i.<br>
a(i) is the average distance between i and all the other data points in the cluster to which i belongs.<br>
b(i) is the average distance from i to all clusters to which i does not belong.

Since we have a large dataset we just check silhouette score for small sample of dataset

```python
#first converting df2 to pandas dataframe in order to make a random sampling
df3 = pd.DataFrame(df2, columns = ['order_dow', 'order_hour_of_day', 'days_since_prior_order', 'add_to_cart_order', 'reordered'])

df3 = df3.sample(frac = 0.02, random_state = 42)
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = []
for num_clusters in K:
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(df3)
    cluster_labels = kmeans.labels_
    
    silhouette_avg.append(silhouette_score(df3, cluster_labels))
```

```python
plt.plot(K,silhouette_avg,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

According two both methods of Elbow and silhouette score we choose 3 as optimal numbe of clusters

**DBSCAN**

In order to implement DBSCAN first we need to acquire optimum parameters value such as epsilon and minPoints number.<br>
To acquire optimum epsilon number we use k-distance graph method.

```python
from sklearn.neighbors import NearestNeighbors
neighbors = NearestNeighbors(n_neighbors=10)
nbrs = neighbors.fit(X)
distances, indices = nbrs.kneighbors(X)
```

```python
# Plotting K-distance Graph
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(20,10))
plt.plot(distances)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```
